```python
from PIL import Image
import cv2
import numpy as np
import tensorflow as tf
from tensorflow import keras

# Load the gender prediction model from the .h5 file
model = keras.models.load_model('models/gender_model.h5')

# Open a connection to the webcam (or video file)
cap = cv2.VideoCapture(0)  # Change to the appropriate camera index or video file path

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()

    # Convert the frame to grayscale
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Resize or preprocess the image as before
    target_size = (100, 100)
    gray_frame = cv2.resize(gray_frame, target_size)

    # Convert the image to a NumPy array using tf.keras
    input_image = tf.keras.preprocessing.image.img_to_array(gray_frame)

    # Expand the dimensions to match the model's input shape
    input_image = tf.expand_dims(input_image, axis=-1)  # Add a channel dimension for grayscale
    input_image = tf.expand_dims(input_image, axis=0)

    # Make predictions using the loaded and preprocessed image
    predictions = model.predict(input_image)

    # Interpret and post-process the predictions
    predicted_class_index = tf.argmax(predictions[0]).numpy()
    predicted_class = "Male" if predicted_class_index == 0 else "Female"

    # Display the result on the frame
    cv2.putText(frame, f'Predicted Gender: {predicted_class}', (10, 30),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)

    # Display the resulting frame
    cv2.imshow('Gender Prediction', frame)

    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close all windows
cap.release()
cv2.destroyAllWindows()
```

```python
from PIL import Image, ImageDraw, ImageFont
import numpy as np
import tensorflow as tf
from tensorflow import keras
import cv2

# Load the age detection model from the .h5 file
model = keras.models.load_model('models/age_model_acc_0.779.h5')

# Define the age classes
age_classes = ['1-2', '3-9', '10-20', '21-27', '28-45', '46-65', '66-116']

# Replace 'path/to/your/image.jpg' with the actual path to your image file
image_path = 'TestImages/1.jpeg'

# Load the image
image = Image.open(image_path)

# Convert the image to grayscale
gray_image = image.convert('L')  # 'L' mode stands for grayscale

# Resize or preprocess the image as before
target_size = (200, 200)  # Adjust according to your model's input size
gray_image = gray_image.resize(target_size)

# Convert the image to a NumPy array using tf.keras
input_image = np.expand_dims(tf.keras.preprocessing.image.img_to_array(gray_image), axis=-1)

# Expand the dimensions to match the model's input shape
input_image = tf.expand_dims(input_image, axis=0)

# Make predictions using the loaded and preprocessed image
predictions = model.predict(input_image)

# Interpret and post-process the predictions
predicted_class_index = tf.argmax(predictions[0]).numpy()
predicted_age_class = age_classes[predicted_class_index]

# Display the image with predicted age class
image_np = np.array(gray_image)
image_with_text = Image.fromarray(image_np)

# Add predicted age class text to the image
draw = ImageDraw.Draw(image_with_text)
font = ImageFont.load_default()
text_color = "white"  # White color (you can customize the color)
draw.text((10, 10), f'Predicted Age Class: {predicted_age_class}', font=font, fill=text_color)

# Convert to BGR for OpenCV display
image_with_text_bgr = cv2.cvtColor(np.array(image_with_text), cv2.COLOR_RGB2BGR)

# Display the image with the predicted age class
cv2.imshow('Age Prediction', image_with_text_bgr)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
from PIL import Image
import cv2
import numpy as np
import tensorflow as tf
from tensorflow import keras

# Load the age detection model from the .h5 file
model = keras.models.load_model('models/age_model_acc_0.779.h5')

# Define the age classes
age_classes = ['1-2', '3-9', '10-20', '21-27', '28-45', '46-65', '21-27']

# Open a connection to the webcam (or video file)
cap = cv2.VideoCapture(0)  # Change to the appropriate camera index or video file path

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()

    # Convert the frame to grayscale
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Resize or preprocess the image as before
    target_size = (200, 200)  # Adjust according to your model's input size
    gray_frame = cv2.resize(gray_frame, target_size)

    # Convert the image to a NumPy array using tf.keras
    input_image = np.expand_dims(gray_frame, axis=-1)  # Add a channel dimension for grayscale
    input_image = tf.convert_to_tensor(input_image, dtype=tf.float32)  # Convert to TensorFlow tensor

    # Expand the dimensions to match the model's input shape
    input_image = tf.expand_dims(input_image, axis=0)

    # Make predictions using the loaded and preprocessed image
    predictions = model.predict(input_image)

    # Interpret and post-process the predictions
    predicted_class_index = tf.argmax(predictions[0]).numpy()
    predicted_age_class = age_classes[predicted_class_index]

    # Display the result on the frame
    cv2.putText(frame, f'Predicted Age Class: {predicted_age_class}', (10, 30),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)

    # Display the resulting frame
    cv2.imshow('Age Prediction', frame)

    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close all windows
cap.release()
cv2.destroyAllWindows()
```
